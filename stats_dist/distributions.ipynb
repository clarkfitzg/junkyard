```python
from IPython.html.widgets import interact
from scipy.stats import *
# This is a personal library- remove the line below or you'll get ImportError
import trix
```

How we (statisticians) think of random variables:
$$
Z \sim N(0, 1)
$$

We can do exactly this in Python.

```python
Z = norm(0, 1)
type(Z)
```

```python
Z.mean()
```

```python
Z.median()
```

```python
Z.std()
```

```python
Z.var()
```

```python
Z.pdf(0.2)
```

```python
Z.cdf(2)
```

```python
Z.ppf(0.975)
```

```python
# centered 95% interval
Z.interval(0.95)
```

```python
Zsample = Z.rvs(1000)
```

```python
norm.fit(Zsample) # Fits MLE!
```

```python
# Calculate the expectation of a function of the random variable

def f(x):
    'Some silly function'
    return sin(0.7 * (x - 4.8)) ** 2 + 43 * x

norm.expect(f)
```

```python
x = linspace(0, 20)

@interact
def plotchi(n = 10):
    y = chi2(n).pdf(x)
    plot(x, y)
```

```python
@interact
def showchi(n = 10):
    trix.plot_rv_cont(chi2(n))
```

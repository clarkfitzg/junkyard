## Z-score

The Z-score scales and translates data. The transformation is given by $Z(x) = \frac{x - \mu}{\sigma}$, where $\mu$ is the mean and $\sigma$ is the standard deviation.

Note that `scipy.stats.zscore` as used below is transforming each column independently.

```python
def ezplot(name, x):
    '''
    plots before and after z score for n x 2 arrays
    '''
    plt.figure()
    plt.title(name, fontsize=30)
    plt.plot(x[:, 0], x[:, 1], marker='o', linestyle='None', alpha=0.4)
    z = stats.zscore(x)
    plt.plot(z[:, 0], z[:, 1], marker='o', linestyle='None')
```

```python
shape = (1000, 2)

x = {'uniform': stats.uniform().rvs(shape), 
     'exponential': stats.expon().rvs(shape),
     'binomial': stats.binom(20, 0.5).rvs(shape),
     }

for name, array in x.items():
    ezplot(name, array)
```

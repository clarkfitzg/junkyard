```python
theta = np.pi / 2

rot = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
print(rot)
invrot = np.linalg.inv(rot)

scalecov = np.array([[5, 0], [0, 1]])
print(scalecov)

cov = reduce(np.dot, [invrot, scalecov, rot])

print(cov)

n = 2000


gaussian = stats.multivariate_normal(mean=[0, 0], cov=cov).rvs(n)
sns.kdeplot(gaussian, color='red')

```
